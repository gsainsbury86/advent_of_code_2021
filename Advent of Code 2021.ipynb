```python
import numpy as np 

def test_bingo_board(df, numbers):
    res = False
    for i, row in df.iterrows():
        res = res or test_bingo_row(row,numbers)
    for i, col in df.transpose().iterrows():
        res = res or test_bingo_row(col,numbers)
    return res
    
def test_bingo_row(row, numbers):
    test = row.isin(numbers)
    res = test.all()
    return res


def score_bingo_board(df, numbers):
    flat = df.unstack().reset_index(drop=True)
    test = np.logical_not(flat.isin(numbers))

    sum_remaining = flat[test].sum()
    return sum_remaining * numbers[-1:]


board_df = pd.DataFrame(np.array([[22, 13, 17, 11,  0], [8, 2, 23,  4, 24], [21,  9, 14, 16,  7],[6, 10,  3, 18, 5],  [1, 12, 20, 15, 19]]))

print(test_bingo_board(board_df,pd.Series([17, 23, 2, 0, 4, 21, 24, 8])))
print(score_bingo_board(board_df,pd.Series([17, 23, 2, 0, 4, 21, 24, 8])))
print(test_bingo_board(board_df,pd.Series([17, 23, 2, 0, 543, 21, 24, 8])))


```

```python
lines = tuple(open('day_12_input.txt', 'r'))
```

```python
import re
import math
def explode(test_explode_str):
    d = 0
    for i, c in enumerate(test_explode_str):
        if c == '[':
            d+=1
        if c ==  ']':
            d-=1
        if d == 5:
            #print(test_explode_str[i+1:])
            lookahead = re.search('(\d+),(\d+)', test_explode_str[i:])
            left = lookahead.group(1)
            right = lookahead.group(2)
            #print(left, right)
            
            # explode right?
            search_from = i + lookahead.span()[1]
            #print(test_explode_str[search_from:])
            
            lookright = re.search('(\d+)', test_explode_str[search_from:])
            out_right = test_explode_str[search_from+1:]
            if not lookright is None:
                update_right = str(int(right) + int(lookright.group(1)))
                out_right = out_right.replace(lookright.group(1),update_right,1)
                #print("explode right", out_right)
        
            # explode left?
            out_left = test_explode_str[:i]
            out_left_backwards = ''.join(reversed(out_left))
            lookleft = re.search('(\d+)', out_left_backwards)
            #print(lookleft)
            if not lookleft is None:
                update_left = ''.join(reversed(str(int(left) + int(''.join(reversed(lookleft.group(1)))))))
                out_left = ''.join(reversed(out_left_backwards.replace(lookleft.group(1),update_left,1)))
                #print("explode left")

            explode_out = out_left + '0' + out_right
            #print("explode", explode_out)
            return explode_out, True
            
    return test_explode_str, False



#explode('[[[[4,0],[5,4]],[[7,0],[15,5]]],[10,[[0,[11,3]],[[6,3],[8,8]]]]]')
```

```python
def reduce(test_reduce_str):
    reduced_str, exploded = explode(test_reduce_str)
    if exploded:
        return reduce(reduced_str)
    
    reduced_str, splitted = split(test_reduce_str)
    if splitted:
        return reduce(reduced_str)

    return test_reduce_str

def sf_add(num_a, num_b):
    added = "[{},{}]".format(num_a,num_b)
    #print("added",added)
    reduced_str = reduce(added)
    #print(reduced_str)
    return reduced_str

mag_pattern = '\[(.*),(.*)\]'
    
def magnitude(test_mag_str):
    try:
        return int(test_mag_str)
    except:
        pass
    return 3*magnitude(test_mag_str[0]) + 2* magnitude(test_mag_str[1])
```

```python
import itertools
import pandas as pd
import numpy as np

distance_dfs = list()
for scanner in scanners:
    distance_df = np.zeros((len(scanner),len(scanner)))
    distance_dfs.append(distance_df)
    #beacon_pairs = list(itertools.combinations(scanner,2))
    beacon_pairs = list((i,j) for ((i,_),(j,_)) in itertools.combinations(enumerate(scanner), 2))
    #print(len(beacon_pairs))
    for beacon_1_i, beacon_2_i in beacon_pairs:
        beacon_1 = scanner[beacon_1_i]
        beacon_2 = scanner[beacon_2_i]
        #print(beacon_1, beacon_2)
        distance = np.linalg.norm(beacon_1 - beacon_2)
        distance_df[beacon_1_i][beacon_2_i] = distance
    
#print(distance_dfs)

counts = np.zeros(distance_dfs[0].shape)
for search_x in range(0, len(distance_dfs[0])):
    for search_y in range(search_x+1,len(distance_dfs[0][search_x])):
        search_distance = distance_dfs[0][search_x][search_y]
        search_df = distance_dfs[1]

        sel = np.where(search_df == search_distance)
        if len(sel[0]) == 1 and len((sel[1]) == 1):
            #print("distance scanner_1 from {} to {} matches distance scanner_2 from {} to {}".format(search_x, search_y, sel[0], sel[1]))
            counts[search_x][sel[0]]+=1
            counts[search_y][sel[1]]+=1
            

for r in counts:
    max_this_row = max(r)
    if max_this_row > 0:
        print(max_this_row, np.argwhere(r == max_this_row))
```
